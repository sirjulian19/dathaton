```python
import pandas as pd
df_demographics=pd.read_csv('dathaton/Dataset/demographics.csv')
df_geographic=pd.read_csv('dathaton/Dataset/geographic.csv')
df_green_trips_new_2=pd.read_csv('dathaton/Dataset/green_trips_new_2.csv')
df_mta_trips=pd.read_csv('dathaton/Dataset/mta_trips.csv')
df_uber_trips_2014=pd.read_csv('dathaton/Dataset/uber_trips_2014.csv')
df_uber_trips_2015=pd.read_csv('dathaton/Dataset/uber_trips_2015.csv')
df_weather=pd.read_csv('dathaton/Dataset/weather.csv')
df_yellow_trips=pd.read_csv('dathaton/Dataset/yellow_trips_new.csv')
df_zones=pd.read_csv('dathaton/Dataset/zones.csv')
```

TRIPS MTA DATA

```python
df_groupby4=df_mta_trips.groupby(
        ['station'],as_index=False
).agg({
    'unit_id':'count',
'new_entries':'count',
    'new_exits':'count',
    'latitude':'mean',
    'longitude':'mean'
}
)
df_groupby4
```

```python
X_4=df_groupby4[['new_entries','new_exits','latitude','longitude']]
y_4=df_groupby4['unit_id']
```

```python
import statsmodels.api as st
X_constant4=st.add_constant(X_4)
lin_reg4=st.OLS(y_4,X_constant4).fit()
lin_reg4.summary()
```

```python
#assumptions
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.stats.api as sts
```

```python
resid_val4=lin_reg4.resid
fitted_val4=lin_reg4.predict()
#valor esperado de media de residuales igual a cero
lin_reg4.resid.mean()
```

```python
#calcular normalidad residuales con shapiro wilk 
from scipy import stats
stats.shapiro(lin_reg4.resid)
st.qqplot(resid_val4,line='s')
```

```python
stats.shapiro(lin_reg4.resid)
```

```python
#linealidad
sns.regplot(x=fitted_val4,y=y_4,lowess=True,line_kws={'color':'red'})
plt.title('fitted con observados')
```

```python
#para homocedasticidad
resid_stand4=lin_reg4.get_influence().resid_studentized_internal
sns.regplot(x=fitted_val4,y=resid_stand4,lowess=True,line_kws={'color':'red'})
plt.grid()
plt.title('fitted con residuals')
```

```python
fig,ax=plt.subplots(figsize=(12,9))
st.graphics.influence_plot(lin_reg4,alpha=0.05,ax=ax,criterion='cooks')
```

```python
#breusch pagan test
bp_test=sts.het_breuschpagan(resid_val4,lin_reg4.model.exog)
print(bp_test)
print("breusch pagan p values= ",bp_test[1])
```

```python
sns.heatmap(X_4.corr(),annot=True)
```

```python
#script multicolinealidad stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=[variance_inflation_factor(X_constant4.values,i) for i in range(X_constant4.shape[1])]
pd.DataFrame({'vif':vif[1:]},index=X_4.columns).T
#para eliminar vif mayores a cinco
```
